This method splits text into individual words or tokens to create a vocabulary. It then counts how often words appear to generate numerical vector representations. Its main flaw is that it completely ignores the semantic meaning of the text.

```python
#designing vocab
from sklearn.feature_extraction.text import CountVectorizer
text = ["It was the best of times", "it was the worst of times", "it was the age of wisdom", "it was the age of foolishness"] 
vectorizer = CountVectorizer()
vectorizer.fit(text)

print(sorted(vectorizer.vocabulary_)) 
```

```python
#transforming documents
vector = vectorizer.transform(text)
print(vector.shape) #dimension
print(vector.toarray()) #given sentence transformation
```

```python
#taking other docs
text2=["the the the times"]
vector=vectorizer.transform(text2)
print(vector.toarray())
```
